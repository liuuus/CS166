```python
import numpy as np
import matplotlib.pyplot as plt

flips = np.array(np.random.uniform(0, 1, size = 30) < 0.6, dtype=int) * 2 - 1
plt.figure(figsize = (10, 4))
plt.plot(np.cumsum(flips))
plt.title('Path of simulation 1')
plt.xlabel('Coin flips')
plt.ylabel('Outcome value')

final_results = []
for i in range(200):
    flips = np.array(np.random.uniform(0, 1, size = 30) < 0.6, dtype=int) * 2 - 1
    last_position = sum(flips)
    final_results.append(last_position)

plt.figure(figsize=(10, 4))
plt.hist(final_results)
plt.axvline(x=6, color='red', label = 'Theoretical mean')
plt.axvline(x=np.mean(final_results), color = 'red', label = 'Experimental mean')
plt.title('Histogram: 200 simulation results')
plt.xlabel('Outcome value')
plt.ylabel('Count')
plt.legend()
```
