#  2. Simulating exponential random variables

```python
import numpy as np
import scipy.stats as sts
rate_parameter = 3 # This is known as lambda in the exponential distribution
distribution = sts.expon(scale=1/rate_parameter)
sample = distribution.rvs(size=1000)
print(sample)
```

```python
print(np.mean(sample))
```

```python
print(np.var(sample))
```

**Do the mean and variance match the theoretical values of the mean and variance of this distribution?**

I think these values match the values of mean and variance from the assignment description since the time (in minutes) it takes passengers to disembark is normally distributed with mean μ=0.03n and standard deviation where is the number of passengers σ=0.01√n. Given that we have a 1000 of samples, these values approximately match.
